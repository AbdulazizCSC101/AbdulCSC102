```python
#Project 1 
from abc import ABC 

class coup_de_ecriva(ABC): 
    
     
    def Fan_Page(self): 
        pass 

class FC_Circok(coup_de_ecriva): 
    def Fan_Page(self,club): 
        print(f"Welcome to {club}") 
        
class Madiba_FC(coup_de_ecriva): 
    def Fan_Page(self,club): 
        print(f"Welcome to {club}") 
        
class Blue_Jay_FC(coup_de_ecriva): 
    def Fan_Page(self,club): 
        print(f"Welcome to {club}") 
        
class TSG_Walker(coup_de_ecriva): 
    def Fan_Page(self,club): 
        print(f"Welcome to {club}") 
        
club = input("What football club do you support? ") 

if club == "Chelsea": 
    s = FC_Circok() 
    s.Fan_Page(club)  
    
elif club == "Arsenal": 
    g = Madiba_FC() 
    g.Fan_Page(club) 
    
elif club == "Man united": 
    h = Blue_Jay_FC() 
    h.Fan_Page(club) 
    
elif club == "Liverpool": 
    j = TSG_Walker() 
    j.Fan_Page(club) 
    
else:
    print("This system does not recognize the club ")

```
